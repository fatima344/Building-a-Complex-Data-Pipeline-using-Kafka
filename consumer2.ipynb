```python
from kafka import KafkaConsumer
from json import dumps,loads
import json
from time import sleep
import pandas as pd
```

```python
consumer = KafkaConsumer('test2', bootstrap_servers=['localhost:9092'], value_deserializer=lambda x: json.loads(x.decode('utf-8')))
```

```python
data = []
```

```python
for message in consumer:
    data.append(message.value)
```

```python
data
```

```python
df = pd.DataFrame(data)
aapl_df = df[['Open:AAPL', 'High:AAPL', 'Low:AAPL', 'Close:AAPL', 'Adj Close:AAPL', 'Volume:AAPL']]
print(aapl_df)
```

```python

positive_fluctuations = 0
negative_fluctuations = 0

for index, row in aapl_df.iterrows():
    diff = row['Close:AAPL'] - row['Open:AAPL']
    
    if diff > 0:
        positive_fluctuations += 1
    elif diff < 0:
        negative_fluctuations += 1

print(f"Positive fluctuations: {positive_fluctuations}, Negative fluctuations: {negative_fluctuations}")

```
